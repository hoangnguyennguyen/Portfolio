```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
mpg_df: pd.DataFrame = sns.load_dataset("mpg")

print(mpg_df.head())
```

### Figure 1: Scatter plot

```python
ax = sns.scatterplot(x="weight", y="mpg", data=mpg_df)
```

### Hexagonal Binning Plots

When there are lots of data points, the plotted points on a scatter plot can end up overlapping, resulting in a messy graph. So we can use hexbin plots as a cleaner graph that's clearer to read.

```python
sns.set(style="ticks")
sns.jointplot(x=mpg_df.weight, y=mpg_df.mpg, kind="hex")
```

### Contour Plots
Another alternative to scatter plots when data points are densely populated in specific regions(s) is a **contour plot**.
Contour plots are commonly used to show the distribution of weather indicators such as temperature, rainfall, and others on maps of geographical regions.

```python
sns.set_style("white")
sns.kdeplot(x=mpg_df.weight, y=mpg_df.mpg, fill=True)
```

## Creating Plots That Present Summary Statistics of Your Data

### Histogram Revistied

```python
# Create Histogram
ax = sns.displot(mpg_df.weight, bins=50, kde=False)

# Add mean and median line
import numpy as np
mean_value = np.mean(mpg_df.weight)
median_value = np.median(mpg_df.weight)

plt.axvline(x=mean_value, color='blue', label='mean')
plt.axvline(x=median_value, color='green', label='median')

plt.legend(loc="upper right")
```

```python
sns.set(style="white")
ax = sns.boxplot(x='model_year', y='mpg', data=mpg_df, hue="model_year", palette='pastel')

ax.legend(loc="upper left", ncol=5)
```
